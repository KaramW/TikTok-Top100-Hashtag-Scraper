```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
htmlFilePath = '/content/drive/MyDrive/Colab Notebooks/Tiktok/tiktok_trending_now.html'
```

```python
#open the file and read the content
with open(htmlFilePath, 'r') as file:
  page_content = file.read()
```

```python
# Create a Beautifullsoup object and specify the parser
soup = BeautifulSoup(page_content, 'html.parser')
```

```python
# Find all the divs that contain the cards
cards = soup.find_all('div', class_="CommonDataList_cardWrapper__kHTJP index-mobile_cardWrapper__Z2o_q")
```

```python
# Funcation to extract data form one html card
def extract_data(card):
    # Extracting the rank
    rank = card.find('span', {'class': 'RankingStatus_rankingIndex__ZMDrH index-mobile_rankingIndex__9mXja'}).text

    # Extracting the hashtag
    hashtag = card.find('span', {'class': 'CardPc_titleText__RYOWo'}).text

    # Extracting the category
    category = card.find('span', {'class': 'i-icon i-icon-tag-one'}).next_sibling

    # Extracting posts and views
    posts, views = card.find_all('span', {'class': 'CardPc_itemValue__XGDmG'})

    return {'Rank': rank, 'Hashtag': hashtag, 'Category': category, 'Posts': posts.text, 'Views': views.text}
```

```python
# Extract data for all cards and save as a list of dictionaries
data = [extract_data(card) for card in cards]
```

```python
# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
# Save the DataFrame as a CSV file
df.to_excel("/content/drive/MyDrive/Colab Notebooks/Tiktok/top_tiktok_hashtags.xlsx", index=False)
```

```python
# Convert the 'Posts' column to numeric
df['Posts'] = df['Posts'].str.replace('K', '000')  # replace 'K' with '000'
df['Posts'] = df['Posts'].str.replace('M', '000000')  # replace 'M' with '000000'
df['Posts'] = pd.to_numeric(df['Posts'])
```

```python
# Sort the DataFrame by the 'Posts' column
df = df.sort_values('Posts', ascending=True)
```

```python
# If you want to save this filtered DataFrame as a new CSV file
df.to_csv("/content/drive/MyDrive/Colab Notebooks/Tiktok/tiktok_hashtags_least_posts.xlsx", index=False)
```
